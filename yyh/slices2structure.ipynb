```python
# /home/perm/workspace/data/MP_xrd_test_with_slices.jsonl 文件每一行的内容示例如下
# {"xrd": "mp-1112207.json", "bandgap": 0.0, "crystal_system": "cubic", "space_group": "Fm-3m", "formation_energy": -2.347, "formula": "K2HgAsF6", "lattice_parameters": [9.2946, 9.2946, 9.2946, 90.0, 90.0, 90.0], "density": 3.868865432, "slices": "K K K K K K K K Hg Hg Hg Hg As As As As F F F F F F F F F F F F F F F F F F F F F F F F 0 39 ooo 0 36 ooo 0 31 ooo 0 32 ooo 0 27 ooo 0 22 ooo 0 29 ooo 0 25 ooo 0 18 ooo 0 16 ooo 0 20 ooo 0 35 ooo 1 17 ++o 1 38 +oo 1 30 o+o 1 26 oo+ 1 37 ooo 1 24 oo+ 1 34 +oo 1 21 +o+ 1 19 o++ 1 28 o+o 1 33 ooo 1 23 ooo 2 35 +oo 2 16 +oo 2 20 +oo 2 30 ooo 2 24 ooo 2 32 ooo 2 19 ooo 2 39 +oo 2 37 ooo 2 29 ooo 2 27 ooo 2 22 ooo 3 33 ooo 3 28 o+o 3 31 o+o 3 21 oo+ 3 18 o++ 3 17 o+o 3 34 ooo 3 25 oo+ 3 26 oo+ 3 38 ooo 3 36 ooo 3 23 ooo 4 16 ++o 4 19 o+o 4 21 +oo 4 38 +oo 4 30 o+o 4 24 ooo 4 35 +oo 4 37 ooo 4 26 ooo 4 29 o+o 4 33 ooo 4 22 ooo 5 25 oo+ 5 18 oo+ 5 27 oo+ 5 20 oo+ 5 28 ooo 5 32 ooo 5 17 ooo 5 36 ooo 5 23 ooo 5 31 ooo 5 39 ooo 5 34 ooo 6 18 o+o 6 16 o+o 6 38 ooo 6 29 o+o 6 25 ooo 6 33 ooo 6 36 ooo 6 21 ooo 6 26 ooo 6 35 ooo 6 31 o+o 6 22 ooo 7 19 oo+ 7 20 +o+ 7 39 +oo 7 34 +oo 7 37 ooo 7 24 oo+ 7 17 +oo 7 27 oo+ 7 30 ooo 7 28 ooo 7 23 ooo 7 32 ooo 8 17 ooo 8 38 o-o 8 39 ooo 8 16 ooo 8 31 ooo 8 30 -oo 9 29 ooo 9 19 ooo 9 27 ooo 9 28 oo- 9 18 ooo 9 26 o-o 10 21 ooo 10 35 ooo 10 34 oo- 10 20 ooo 10 25 ooo 10 24 -oo 11 37 ooo 11 23 ooo 11 33 ooo 11 32 ooo 11 22 ooo 11 36 ooo 12 19 -oo 12 16 ooo 12 21 o-o 12 20 ooo 12 17 oo- 12 18 ooo 13 27 ooo 13 22 ooo 13 25 ooo 13 23 oo- 13 24 ooo 13 26 ooo 14 29 ooo 14 31 ooo 14 33 o-o 14 32 ooo 14 28 ooo 14 30 ooo 15 36 ooo 15 39 ooo 15 34 ooo 15 35 ooo 15 37 -oo 15 38 ooo "}

# -*- coding: utf-8 -*-
# Hang Xiao 2023.04
# xiaohang07@live.cn
import os,sys,json,gc
from slices.core import SLICES
from pymatgen.core.structure import Structure
import configparser
import time
from tqdm import tqdm
import traceback

os.environ["OMP_NUM_THREADS"] = "1"

def convert_slices_to_structure_add_to_jsonl(input_file, output_file, cif_dir): 
    """
    请逐行提取对应的 SLICES 信息并并转换为 Structure 对象，将结果存储在 structures_converted 列表中，并将结果与对应cif文件中的结构进行比较，计算匹配率：
    1. 读取 input_file 中的每一行 JSONL 记录。
    2. 对于每一行，提取 'slices' 字段并使用 SLICES 类将其转换为 Structure 对象。
    3. 将转换后的 Structure 对象存储在 structures_converted 列表中。
    4. 对于每个 Structure 对象，读取对应的 cif 文件并将其转换为 Structure 对象。
    5. 比较转换后的 structures_converted 对象与 cif 文件中的 Structure 对象，计算匹配率。

    参数：
    - input_file: 包含 SLICES 信息的 JSONL 文件路径。
    - output_file: 输出包含 Structure 信息的 JSONL 文件路径。
    - cif_dir: 存储 cif 文件的目录路径。
    """
    
    config = configparser.ConfigParser()
    config.read('/crystal/yyh/settings.ini') #path of your .ini file
    bond_scaling = config.getfloat("Settings","bond_scaling") 
    delta_theta = config.getfloat("Settings","delta_theta") 
    delta_x = config.getfloat("Settings","delta_x") 
    lattice_shrink = config.getfloat("Settings","lattice_shrink") 
    lattice_expand = config.getfloat("Settings","lattice_expand") 
    angle_weight = config.getfloat("Settings","angle_weight") 
    vbond_param_ave_covered = config.getfloat("Settings","vbond_param_ave_covered") 
    vbond_param_ave = config.getfloat("Settings","vbond_param_ave") 
    repul = config.getboolean("Settings","repul") 
    graph_method = config.get("Settings","graph_method")
    # print(delta_theta,lattice_expand)

    # 依次print参数以确认读取正确；先打印参数名称，再打印对应的值
    print("bond_scaling:", bond_scaling)
    print("delta_theta:", delta_theta)
    print("delta_x:", delta_x)
    print("lattice_shrink:", lattice_shrink)
    print("lattice_expand:", lattice_expand)
    print("angle_weight:", angle_weight)
    print("vbond_param_ave_covered:", vbond_param_ave_covered)
    print("vbond_param_ave:", vbond_param_ave)
    print("repul:", repul)
    print("graph_method:", graph_method)
    
    
    
    check=False # whether to check the results by matching with original structures
    CG=SLICES(graph_method=graph_method, check_results=check, relax_model="m3gnet")
    wrong_cif_lines = [0]
    structures_converted = []
    with open(input_file, 'r', encoding='utf-8') as fin, \
         open(output_file, 'a', encoding='utf-8') as fout:
             
        print(f"开始处理文件: {input_file}")
            
        for i,line in enumerate(tqdm(fin, desc=f'Processing {os.path.basename(input_file)}')):
            print(f"Processing line {i+1}")
            print("line",line)
            record = json.loads(line.strip())

            if i+1 <= max(wrong_cif_lines):
                print(f"跳过已完成: {i+1}")
                continue
            
            xrd_file = record.get('xrd')
            if not xrd_file:
                print(f"未找到xrd字段")
                # 报错，停止运行
                print(f"错误行号: {i+1}")
                raise ValueError("xrd字段缺失")
                break

            cif_name = xrd_file.replace('.json', '.cif')
            cif_path = os.path.join(cif_dir, cif_name)
            if not os.path.isfile(cif_path):
                print(f"未找到cif文件: {cif_path}")
                # 报错，停止运行
                print(f"错误行号: {i+1}")
                raise ValueError("cif文件缺失")
                break
            
            try:
                ori = Structure.from_file(cif_path)                
                num_atoms = len(ori.atomic_numbers)
            except Exception as e:
                print(f"处理 {cif_path} 时出错: {e}")
                # 报错，停止运行
                print(f"错误行号: {i+1}")
                wrong_cif_lines.append(i+1)
                break
            
            # print(f"已经读取 CIF file: {cif_path}")
            
            slices_str = record.get('slices')
            if not slices_str:
                print(f"未找到slices字段")
                # 报错，停止运行
                print(f"错误行号: {i+1}")
                raise ValueError("slices字段缺失")
                break
            
            # print(f"准备从 SLICES 转换为 Structure")
            
            try:
                start_time = time.time()
                # CG.from_cif(cif_path) # from_cif 方法内部会产生 SLICES 结构
                #print(bond_scaling, delta_theta, delta_x,lattice_shrink,lattice_expand,angle_weight,epsilon,repul)
                structures, energy=CG.to_structures(bond_scaling, delta_theta, delta_x, \
                lattice_shrink,lattice_expand,angle_weight,vbond_param_ave_covered,vbond_param_ave,repul)
                # print(slices_str)
                # structures, energy = CG.SLICES2structure(slices_str)
                print("energy: ",energy)
                print("structures: ",structures)
                structures_converted.extend(structures)
                time_used=(time.time() - start_time)
                print("num_atoms: ", num_atoms, "time_used: ", time_used)
                # print(f"准备与读取的 CIF file 做结构比较: {cif_path}")
                if len(structures)==3:
                    print("3 structures generated")
                    a,b,c,d,e,f=CG.match_check3(ori,structures[2],structures[1],structures[0])
                    a2,b2,c2,d2,e2,f2=CG.match_check3(ori,structures[2],structures[1],structures[0],ltol=0.3, stol=0.5, angle_tol=10)
                    time_used_compare=(time.time() - start_time)-time_used
                    fout.write(record["xrd"]+','+a+','+b+','+c+','+a2+','+b2+','+c2+','+d2+','+e2+','+f2+','+str(num_atoms)+','+str(time_used)+','+str(time_used_compare)+'\n')
                    print("num_atoms: ", num_atoms, "time_used_compare: ", time_used_compare)
                elif len(structures)==2:
                    print("2 structures generated")
                    a,b,c,d=CG.match_check(ori,structures[1],structures[0])
                    a2,b2,c2,d2=CG.match_check(ori,structures[1],structures[0],ltol=0.3, stol=0.5, angle_tol=10)
                    time_used_compare=(time.time() - start_time)-time_used
                    fout.write(record["xrd"]+','+'0'+','+a+','+b+','+'0'+','+a2+','+b2+','+'1'+','+c2+','+d2+','+str(num_atoms)+','+str(time_used)+','+str(time_used_compare)+'\n')
                    print("num_atoms: ", num_atoms, "time_used_compare: ", time_used_compare)
                else:
                    print("len(structures): ",len(structures))
                    time_used_compare=(time.time() - start_time)-time_used
                    fout.write(record["xrd"]+','+'0'+','+'0'+','+'0'+','+'0'+','+'0'+','+'0'+','+'0'+','+'0'+','+'0'+','+str(num_atoms)+','+str(time_used)+','+str(time_used_compare)+'\n')
                fout.flush()
            except Exception as e1:
                print(f"转换 SLICES 为 Structure 时出错: {e1}")
                print(f"错误行号: {i+1}")
                traceback.print_exc()  # 输出完整堆栈信息
                wrong_cif_lines.append(i+1)
                del CG
                CG=SLICES(graph_method=graph_method, check_results=check, relax_model="m3gnet")
                fout.write(record["xrd"]+','+str(e1).split('\n')[0]+'\n')
                continue

            fout.write(json.dumps(record, ensure_ascii=False) + '\n')

    # 输出所有错误行号
    if wrong_cif_lines:
        # print("nothing")
        print(f"以下行的cif文件缺失: {wrong_cif_lines}")

# 用法示例
if __name__ == "__main__":
    savepath = '/home/perm/workspace/data/'
    cif_dir = '/mnt/minio/battery/xrd/datasets/raw_data/mp_cif/'  # 假设cif文件都在这个目录
    input_file = savepath + 'MP_xrd_test_with_slices.jsonl'
    output_file = savepath + 'results.csv'
    convert_slices_to_structure_add_to_jsonl(input_file, output_file, cif_dir)
```
