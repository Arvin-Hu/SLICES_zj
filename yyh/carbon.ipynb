```python
from yyh.CC import graphene_generator

# 创建生成器
generator = graphene_generator.BilayerTwistedGraphene()

# 生成特定结构
generator.generate_structure(angle=5.0, lx=100, ly=100, lz=15)

# 可视化
generator.visualize()

# 保存文件（会自动生成.lmp数据和.in输入脚本）
generator.save_lammps_data("yyh/CC/TBLG_{generator.angle}deg.lmp")

# 运行模拟：
# mpirun -np 4 lammps -in twisted_graphene_input.in -log simulation.log
```

```python
from yyh.CC import GFNFF

lmp_file = "yyh/CC/TBLG_5.0deg.lmp"
xyz_file = "TBLG_5.0deg.xyz"
out_file = "TBLG_5.0deg_optimize.GFNFF"
GFNFF.lmp_to_xyz(lmp_file, xyz_file)
GFNFF.optimize_with_xtb(xyz_file, out_file)
print(f"优化完成，结果已保存到 {out_file}")
```
